```python
from collections import deque
```

```python
def is_bipartite(graph):
    nodes = set(graph.keys())
    for nbrs in graph.values():
        for v in nbrs:
            nodes.add(v)
    colors = {n: 0 for n in nodes}

    for start in nodes:
        if colors[start] != 0:
            continue
        colors[start] = 1
        q = deque([start])
        while q:
            v = q.popleft()
            for u in graph.get(v, []):
                if colors[u] == 0:
                    colors[u] = -colors[v]
                    q.append(u)
                elif colors[u] == colors[v]:
                    return False, colors
    return True, colors
```

```python
g1 = {1:[2], 2:[1,3], 3:[2]}
ok, cols = is_bipartite(g1)
assert ok == True
assert set(cols[n] for n in cols) <= {1, -1, 0}

g_odd_cycle = {1:[2,3], 2:[1,3], 3:[1,2]}
ok2, cols2 = is_bipartite(g_odd_cycle)
assert ok2 == False

g_mix = {1:[2], 2:[1], 3:[4,5], 4:[3,5], 5:[3,4]}
ok3, _ = is_bipartite(g_mix)
assert ok3 == False

ok_empty, cols_empty = is_bipartite({})
assert ok_empty == True
assert cols_empty == {}

print("Все тесты пройдены успешно")
```
