```python
def coloring_connectivity_components(graph):
    nodes = set(graph.keys())
    for nbrs in graph.values():
        for v in nbrs:
            nodes.add(v)
    visited = {}
    color = 0
    for start in nodes:
        if start in visited:
            continue
        color += 1
        stack = [start]
        visited[start] = color
        while stack:
            v = stack.pop()
            for u in graph.get(v, []):
                if u not in visited:
                    visited[u] = color
                    stack.append(u)
    return visited
```

```python
graph_sem = {
    1: [2,3],
    2: [1,3],
    3: [1,2],
    4: [6,7],
    5: [6,7],
    6: [4,5,7],
    7: [4,5,6],
    8: [11],
    9: [10,11],
    10: [9],
    11: [8,9]
}

colors = coloring_connectivity_components(graph_sem)
comp1 = {n for n,c in colors.items() if c == colors[1]}
comp2 = {n for n,c in colors.items() if c == colors[4]}
comp3 = {n for n,c in colors.items() if c == colors[8]}

assert comp1 == {1,2,3}
assert comp2 == {4,5,6,7}
assert comp3 == {8,9,10,11}

print("Пример из семинара - раскраска компонент (node:color):", colors)

assert coloring_connectivity_components({}) == {}
assert coloring_connectivity_components({1: []}) == {1: 1}

print("Все тесты пройдены успешно")
```
