```python
class Solution:
    def findTheDifference(self, s: str, t: str) -> str:
        count = {}

        for ch in s:
            count[ch] = count.get(ch, 0) + 1

        for ch in t:
            if ch not in count or count[ch] == 0:
                return ch
            count[ch] -= 1
```

```python
sol = Solution()

tests = [
    ("basic 1", sol.findTheDifference("abcd", "abcde"), "e"),
    ("basic 2", sol.findTheDifference("xyz", "wxyz"), "w"),
    ("basic 3", sol.findTheDifference("ace", "aecb"), "b"),

    ("duplicates", sol.findTheDifference("aabbc", "ababcx"), "x"),
    ("duplicate added", sol.findTheDifference("aabb", "ababb"), "b"),

    ("empty s", sol.findTheDifference("", "q"), "q"),
    ("many same", sol.findTheDifference("aaaaa", "aaaaab"), "b"),

    ("permutation", sol.findTheDifference("qwerty", "ytwerqx"), "x"),

    ("spaces", sol.findTheDifference("ab ", "a b "), " "),
    ("unicode", sol.findTheDifference("абвг", "вгабд"), "д"),

    ("long", sol.findTheDifference("a"*100000, "a"*100000 + "b"), "b"),
]

for name, got, expected in tests:
    print(name, "OK" if got == expected else f"FAIL (got {got}, expected {expected})")
```
