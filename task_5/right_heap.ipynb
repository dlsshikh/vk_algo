```python
import heapq
from collections import deque

class Tree_Node:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def arr_to_bin_tree(arr, i=0):
    if i >= len(arr) or arr[i] is None:
        return None

    node = Tree_Node(arr[i])
    node.left = arr_to_bin_tree(arr, 2*i + 1)
    node.right = arr_to_bin_tree(arr, 2*i + 2)
    return node
```

```python
def is_max_heap_array(arr):
    n = len(arr)
    for i in range((n - 2) // 2 + 1):
        left = 2 * i + 1
        right = 2 * i + 2
        
        if left < n and arr[i] < arr[left]:
            return False
        if right < n and arr[i] < arr[right]:
            return False
    return True
```

```python
def is_min_heap_array(arr):
    n = len(arr)
    for i in range((n - 2) // 2 + 1):
        left = 2 * i + 1
        right = 2 * i + 2
        
        if left < n and arr[i] > arr[left]:
            return False
        if right < n and arr[i] > arr[right]:
            return False
    return True
```

```python
def is_max_heap_bfs(root):
    if not root:
        return True
    
    queue = deque([root])
    while queue:
        current = queue.popleft()
        
        if current.left:
            if current.left.val > current.val:
                return False
            queue.append(current.left)
        if current.right:
            if current.right.val > current.val:
                return False
            queue.append(current.right)
    return True
```

```python
max_heap_arr = [21, 19, 18, 11, 12, 15, 16, 9, 8, 10]
min_heap_arr = [1, 2, 3, 4, 5, 6, 7]
tree1 = arr_to_bin_tree(max_heap_arr)
tree2 = arr_to_bin_tree(min_heap_arr)

print(f"Max heap check: {is_max_heap_array(max_heap_arr)}")
print(f"Max heap check (BFS): {is_max_heap_bfs(tree1)}")
print(f"Max heap check (BFS): {is_max_heap_bfs(tree2)}")
print(f"Min heap check: {is_min_heap_array(min_heap_arr)}")
```
