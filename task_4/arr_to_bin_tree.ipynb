Восстановление бинарного дерева из массива

```python
class Tree_Node:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def arr_to_bin_tree( arr, i=0):
    if i >= len(arr) or arr[i] is None:
        return None

    node = Tree_Node(arr[i])
    node.left = arr_to_bin_tree(arr, 2*i + 1)
    node.right = arr_to_bin_tree(arr, 2*i + 2)
    return node
```

```python
arr = [8, 9, 11, 7, 16, 3, 1]
tree = arr_to_bin_tree(arr)
tree
```
