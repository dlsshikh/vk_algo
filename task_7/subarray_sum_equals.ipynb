```python
from collections import defaultdict
```

```python
def subarray_sum_equals(nums, k):
    prefix_sum = 0
    count = 0
    prefix_counts = defaultdict(int)
    prefix_counts[0] = 1
    for x in nums:
        prefix_sum += x
        count += prefix_counts[prefix_sum - k]
        prefix_counts[prefix_sum] += 1
    return count
```

```python
a = [3, 8, 6, 9, 2, 1, 4]
k = 11
print(subarray_sum_equals(a, k), " expected 2 (пример из слайдов)")

print(subarray_sum_equals([1,1,1], 2), " expected 2")
print(subarray_sum_equals([], 0), " expected 0")
print(subarray_sum_equals([0,0,0], 0), " expected 6")
print(subarray_sum_equals([1,-1,0], 0), " expected 3")
```
