```python
def find_max_length_equal_zeros_ones(nums):
    prefix_sum = 0
    index_map = {0: -1}
    max_len = 0
    for i, val in enumerate(nums):
        if val == 0:
            prefix_sum -= 1
        else:
            prefix_sum += 1
        if prefix_sum in index_map:
            max_len = max(max_len, i - index_map[prefix_sum])
        else:
            index_map[prefix_sum] = i
    return max_len
```

```python
print(find_max_length_equal_zeros_ones([0, 1, 0, 0, 1]), " expected 4")
print(find_max_length_equal_zeros_ones([0,0,0,0]), " expected 0")
print(find_max_length_equal_zeros_ones([1,1,0,0,1,0]), " expected 6")
print(find_max_length_equal_zeros_ones([]), " expected 0")
print(find_max_length_equal_zeros_ones([0,1]), " expected 2")
```
