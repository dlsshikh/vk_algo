```python
def max_subarray_sum(arr, k):
    n = len(arr)
    if k > n or k <= 0:
        return None

    current_sum = sum(arr[:k])
    max_sum = current_sum

    for i in range(k, n):
        current_sum += arr[i] - arr[i - k]
        if current_sum > max_sum:
            max_sum = current_sum

    return max_sum
```

```python
arr = [2, 1, 5, 1, 3, 2]
print(max_subarray_sum(arr, 3), " expected 9")

print(max_subarray_sum([], 1), " expected None (пустой массив)")
print(max_subarray_sum([1,2], 3), " expected None (k > n)")
print(max_subarray_sum([-2, -1, -3, -4], 2), " expected -3")
print(max_subarray_sum([5], 1), " expected 5")
```
