```python
class Solution:
    def Shell_sort(self, array):
        gap = len(array)//2
        while gap > 0:
            for i in range(0, len(array) - gap):
                if array[i] > array[i + gap]:
                    array[i], array[i + gap] = array[i + gap], array[i]
            gap = gap // 2
        return array
            
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        cache = {}
        for i, num in enumerate(nums):
            diff = target - num #проверяем, соответствует ли разница второму элементу
            if diff in cache:
                return [cache[diff], i]
            cache[num] = i
        return 'did not found :('
```

```python
sol = Solution()

tests = [
    ("basic 1", sol.twoSum([2,7,11,15], 9), [0,1]),
    ("basic 2", sol.twoSum([3,2,4], 6), [1,2]),
    ("basic 3", sol.twoSum([3,3], 6), [0,1])]

for name, got, expected in tests:
    print(name, "OK" if got == expected else f"FAIL (got {got}, expected {expected})")
```
