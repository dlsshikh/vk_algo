```python
def has_cycle_undirected(graph):
    nodes = set(graph.keys())
    for nbrs in graph.values():
        for v in nbrs:
            nodes.add(v)
    visited = set()

    def dfs(v, parent) -> bool:
        visited.add(v)
        for u in graph.get(v, []):
            if u == v:
                return True
            if u not in visited:
                if dfs(u, v):
                    return True
            elif u != parent:
                return True
        return False
    for node in nodes:
        if node not in visited:
            if dfs(node, None):
                return True
    return False
```

```python
g_cycle = {1:[2,3], 2:[1,3], 3:[1,2]}

graph_sem = {
    1: [2,3],
    2: [1,3],
    3: [1,2],
    4: [6,7],
    5: [6,7],
    6: [4,5,7],
    7: [4,5,6],
    8: [11],
    9: [10,11],
    10: [9],
    11: [8,9]
}
assert has_cycle_undirected(g_cycle) == True

g_tree = {1:[2,3], 2:[1,4], 3:[1], 4:[2]}
assert has_cycle_undirected(g_tree) == False

assert has_cycle_undirected(graph_sem) == True

assert has_cycle_undirected({1:[1]}) == True

assert has_cycle_undirected({}) == False
assert has_cycle_undirected({1:[]}) == False

print("Все тесты пройдены успешно")
```
