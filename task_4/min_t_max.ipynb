```python
class Tree_Node:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
def arr_to_bin_tree( arr, i=0):
    if i >= len(arr) or arr[i] is None:
        return None

    node = Tree_Node(arr[i])
    node.left = arr_to_bin_tree(arr, 2*i + 1)
    node.right = arr_to_bin_tree(arr, 2*i + 2)
    return node
```

```python
def max_min_multiplication_complete(root):
    """Поиск min и max через обход всего дерева"""
    if not root:
        return -1
    
    def find_min_max(node, current_min, current_max):
        if not node:
            return current_min, current_max
        
        current_min = min(current_min, node.val)
        current_max = max(current_max, node.val)
        
        left_min, left_max = find_min_max(node.left, current_min, current_max)
        right_min, right_max = find_min_max(node.right, current_min, current_max)
        
        return min(current_min, left_min, right_min), max(current_max, left_max, right_max)
    
    min_val, max_val = find_min_max(root, float('inf'), float('-inf'))
    return min_val * max_val
```
