Как быстро можно сделать N копий документа, используя два ксерокса, каждый копирует со своей скоростью (x и y минут)?

```python
class Solution:
    def minimumTime(self, n: int, x_time: int, y_time: int) -> int:
        first = min(x_time, y_time)
        n-=1
        if n <= 0:
            return first
        l = 0
        r = n * min(x_time, y_time)
        while l < r:
            mid = (l + r) // 2
            if mid//x_time + mid//y_time < n: #время слишком маленькое,   
                l = mid + 1    #какой-то принтер перегружен (времена накладываются)
            else: #время слишком большое,
                r = mid                            #какой-то принтер недозагружен (есть свободные окна, в которые можно напечатать)
        return l + first
```

```python
#один ксерокс быстрее
print(Solution().minimumTime(4, 1, 2))  
# Ожидается: 3
# 1 минута — первая копия на быстром, 
# за следующие 2 минуты оба делают оставшиеся 3 копии → всего 3 мин.

#Равные скорости
print(Solution().minimumTime(5, 2, 2))  
# Ожидается: 6
# 1-я копия за 2 мин, остаётся 4; оба по 1 копии каждые 2 мин.
# 4 копии / 2 ксерокса = 2 раунда по 2 мин → 2 + 2*2 = 6 мин

#Очень разные скорости
print(Solution().minimumTime(5, 1, 10))  
# Ожидается: 5
# Первый делает копии почти один: 1-я за 1 мин, потом за 4 мин ещё 4 копии = 5 мин.

#Минимальный случай
print(Solution().minimumTime(1, 5, 10))  
# Ожидается: 5
# Нужно всего 1 копию, берём время самого быстрого аппарата.

#Большое N (проверка производительности)
print(Solution().minimumTime(1000000, 1, 2))  
# Ожидается: около 666667
# Первый делает 1 копию за 1 мин, потом оба работают параллельно:
# оптимальное время ≈ n / (1/1 + 1/2) = n * 2/3 ≈ 666667
```
