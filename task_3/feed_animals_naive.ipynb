В небольшом зоопарке есть некоторое количество животных.

Каждое животное потребляет какой-то объем еды, выраженный в целочисленном значении. 

Например, еноту нужна 1-порция еды, зебре 2, пантере 3, льву 4, жирафу 8 и т.д.

Каждый день, смотритель зоопарка привозит еду такими же порциями. 

То есть за раз он привозит 8, 3, 9, 1, 7. Порция на 8 может накормить одно животное один раз. 

То есть такая порция может накормить либо енота, либо льва, либо жирафа, но не может накормить, например зебру и енота. Только кого-то одного.

Надо написать функцию, которая определит, сколько из переданных животных может накормить заданное количество еды.

```python
class Solution:
    def Feed_Animals(self, g: list[int], s: list[int]) -> int:
        #граничный случай
        if len(g)==0 and len(s)==0:
            return ('Животные всё скушали и сбежали из зоопарка :(')
        else:
            counter = 0 #индекс, по которому обращаемся к животным
            for size in s:
                for animal in g: 
                    if len(g) > counter: #пока не накормили всех
                        if size >= g[counter]:
                            counter +=1
                if len(g) == counter:
                    break
        
        return counter        
```

```python
sol = Solution()

tests = [
    ("basic 1", sol.Feed_Animals([1,2], [1,2,3]), 2),
    ("basic 2", sol.Feed_Animals([1,2,3], [1,1]), 1),
    ("basic 3", sol.Feed_Animals([2,3,4], [1,2,3]), 2),
    ("empty both", sol.Feed_Animals([], []), 'Животные всё скушали и сбежали из зоопарка :('),
    ("empty animals", sol.Feed_Animals([], [1,2,3]), 0),
    ("empty food", sol.Feed_Animals([1,2,3], []), 0),
    ("exact", sol.Feed_Animals([1,2,3], [1,2,3]), 3),
    ("more food", sol.Feed_Animals([2,3], [5,5,5,5]), 2),
    ("more animals", sol.Feed_Animals([1,2,3,4], [3,3]), 2),
    ("unsorted", sol.Feed_Animals([3,1,2], [2,3,1]), 3),
    ("repeating", sol.Feed_Animals([1,1,2,2], [1,1,1]), 2),
    ("big sparse", sol.Feed_Animals([10,20,30], [5,25,35]), 2),
]

for name, got, expected in tests:
    print(name, "OK" if got == expected else f"FAIL (got {got}, expected {expected})")
```
