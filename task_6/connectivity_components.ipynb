```python
from collections import deque
```

```python
def find_connectivity_components(graph):
    nodes = set(graph.keys())
    for nbrs in graph.values():
        for v in nbrs:
            nodes.add(v)
    visited = set()
    components = []
    for start in nodes:
        if start in visited:
            continue
        comp = []
        stack = [start]
        visited.add(start)
        while stack:
            v = stack.pop()
            comp.append(v)
            for u in graph.get(v, []):
                if u not in visited:
                    visited.add(u)
                    stack.append(u)
        components.append(comp)
    return components
```

```python
graph_sem = {
    1: [2,3],
    2: [1,3],
    3: [1,2],
    4: [6,7],
    5: [6,7],
    6: [4,5,7],
    7: [4,5,6],
    8: [11],
    9: [10,11],
    10: [9],
    11: [8,9]
}
comps = find_connectivity_components(graph_sem)
normalized = [set(c) for c in comps]
assert any(set([1,2,3]) == s for s in normalized)
assert any(set([4,5,6,7]) == s for s in normalized)
assert any(set([8,9,10,11]) == s for s in normalized)
print("Пример из семинара - компоненты:", comps)

assert find_connectivity_components({}) == []
assert find_connectivity_components({1: []}) == [[1]]
assert len(find_connectivity_components({1:[2,3],2:[1,3],3:[1,2]})) == 1

print('Все тесты успешно пройдены')
```
