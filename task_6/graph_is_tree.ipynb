```python
def has_cycle_undirected(graph):
    nodes = set(graph.keys())
    for nbrs in graph.values():
        for v in nbrs:
            nodes.add(v)
    visited = set()

    def dfs(v, parent) -> bool:
        visited.add(v)
        for u in graph.get(v, []):
            if u == v:
                return True
            if u not in visited:
                if dfs(u, v):
                    return True
            elif u != parent:
                return True
        return False
    for node in nodes:
        if node not in visited:
            if dfs(node, None):
                return True
    return False
```

```python
def find_connectivity_components(graph):
    nodes = set(graph.keys())
    for nbrs in graph.values():
        for v in nbrs:
            nodes.add(v)
    visited = set()
    components = []
    for start in nodes:
        if start in visited:
            continue
        comp = []
        stack = [start]
        visited.add(start)
        while stack:
            v = stack.pop()
            comp.append(v)
            for u in graph.get(v, []):
                if u not in visited:
                    visited.add(u)
                    stack.append(u)
        components.append(comp)
    return components
```

```python
def graph_is_tree(graph):
    nodes = set(graph.keys())
    for nbrs in graph.values():
        for v in nbrs:
            nodes.add(v)
    if len(nodes) == 0:
        return False
    if has_cycle_undirected(graph):
        return False
    comps = find_connectivity_components(graph)
    return len(comps) == 1
```

```python
g_tree = {1:[2,3], 2:[1,4], 3:[1], 4:[2]}
assert graph_is_tree(g_tree) == True

g_cycle = {1:[2,3], 2:[1,3], 3:[1,2]}
assert graph_is_tree(g_cycle) == False

assert graph_is_tree({1:[], 2:[]}) == False

assert graph_is_tree({1: []}) == True

assert graph_is_tree({}) == False

print("Все тесты пройдены успешно")
```
