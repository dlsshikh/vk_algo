```python
'''Найти корень числа (ближайшее целое)'''
```

```python
class Solution:
    def mySqrt(self, x: int) -> int:
        if x < 2:
            return x

        l = 0
        r = x
        while l <= r:
            middle = (l+r) // 2
            if middle*middle > x:
                r = middle - 1
            elif middle*middle < x:
                l = middle + 1
            elif middle*middle == x:
                return middle
        
        return r
```
