```python
def pivot_index(nums):
    total = sum(nums)
    left_sum = 0
    for i, val in enumerate(nums):
        if left_sum == total - left_sum - val:
            return i
        left_sum += val
    return -1
```

```python
print(pivot_index([1, 7, 3, 6, 5, 6]), " expected 3")
print(pivot_index([1,2,3]), " expected -1")
print(pivot_index([2,1,-1]), " expected 0")
print(pivot_index([],), " expected -1")
print(pivot_index([0,0,0]), " expected 0")
```
