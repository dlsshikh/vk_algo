```python
class Solution:
    def groupAnagrams(self, strs: list[str]) -> list[list[str]]:
        anagrams = {}  # ключ = отсортированная строка, значение = список слов

        for word in strs:
            key = ''.join(sorted(word))  # сортируем буквы в слове
            if key not in anagrams:
                anagrams[key] = []
            anagrams[key].append(word)

        return list(anagrams.values())
```

```python
sol = Solution()

tests = [
    ("basic 1", sol.groupAnagrams(["eat","tea","tan","ate","nat","bat"]), [['eat', 'tea', 'ate'], ['tan', 'nat'], ['bat']]),
    ("basic 2", sol.groupAnagrams([""]), [[""]]),
    ("basic 3", sol.groupAnagrams(["a"]), [["a"]])]

for name, got, expected in tests:
    print(name, "OK" if got == expected else f"FAIL (got {got}, expected {expected})")
```
