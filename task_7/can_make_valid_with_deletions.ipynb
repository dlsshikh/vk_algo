```python
def can_make_valid_with_deletions(s, k):
    balance = 0
    extra_closed = 0
    for ch in s:
        if ch == '(':
            balance += 1
        elif ch == ')':
            if balance > 0:
                balance -= 1
            else:
                extra_closed += 1
        else:
            pass
    total_needed = balance + extra_closed
    return total_needed <= k
```

```python
tests = [
    ("()", 0, True),
    ("(()", 1, True),
    ("))()", 2, True),
    ("()))", 2, True),
    ("() ((", 3, False),
    (") )", 2, True),
    ("()(( ", 3, False),
]
print(can_make_valid_with_deletions("()", 0), " expected True")
print(can_make_valid_with_deletions("(()", 1), " expected True")
print(can_make_valid_with_deletions("))()", 2), " expected True")
print(can_make_valid_with_deletions("()))", 2), " expected True")
print(can_make_valid_with_deletions("()(((", 3), " expected True (удалить три '(')")
print(can_make_valid_with_deletions("", 0), " expected True (пустая строка уже валидна)")
print(can_make_valid_with_deletions("))))", 3), " expected False (нужно 4 удалений)")
print(can_make_valid_with_deletions("((((((", 6), " expected True (удалить все 6)")
```
