```python
class Tree_Node:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
def arr_to_bin_tree( arr, i=0):
    if i >= len(arr) or arr[i] is None:
        return None

    node = Tree_Node(arr[i])
    node.left = arr_to_bin_tree(arr, 2*i + 1)
    node.right = arr_to_bin_tree(arr, 2*i + 2)
    return node
```

Проверка на симметричность через обход в ширину

```python
from collections import deque

def isSymmetric_bfs(root):
    if not root:
        return True
    
    queue = deque([root])
    
    while queue:
        level_size = len(queue)
        level_nodes = []
        
        for _ in range(level_size):
            node = queue.popleft()
            if node:
                level_nodes.append(node.val)
                queue.append(node.left)
                queue.append(node.right)
            else:
                level_nodes.append(None)
        
        # Проверяем, является ли уровень палиндромом
        left, right = 0, len(level_nodes) - 1
        while left < right:
            if level_nodes[left] != level_nodes[right]:
                return False
            left += 1
            right -= 1
    
    return True
```

Проверка на симметричность через обход в глубину

```python
def isSymmetric_dfs(root):
    if not root:
        return True
    
    def inorder_traversal(node, result):
        if not node:
            return
        inorder_traversal(node.left, result)
        result.append(node.val)
        inorder_traversal(node.right, result)
    
    data = []
    inorder_traversal(root, data)
    
    # Проверяем, является ли массив симметричным
    j = len(data) - 1
    for i in range(len(data) // 2):
        if data[i] != data[j]:
            return False
        j -= 1
    
    return True

'''Альтернативный рекурсивный подход DFS
def isSymmetric_recursive(root):
    def is_mirror(left, right):
        if not left and not right:
            return True
        if not left or not right:
            return False
        return (left.val == right.val and 
                is_mirror(left.left, right.right) and 
                is_mirror(left.right, right.left))
    
    return is_mirror(root, root) if root else True'''
```
